```python
h=0.1

x= np.arange(0, 2*np.pi, h) #arrange = 0 - 2phi -> pi yang dipakai adalah 3.14

# compute function
y= np.cos(x) #fungsi cos(x) adalah fungsi sebagai soal

# compute vector of foward differences
# rumus dasar f(xi+h)-f(x) /h
# jd jk fungsi ini dipakai dalam foward diff --> cos(xi+h)-cos(x) / h
foward_diff = np.diff(y)/h

# compute corresponding grid
x_diff = x[:-1:] #dihitung sampai 6.1 krn di -1 dari 6.2

# compute exact solution (merupakan hasil turunan dari fungsi cos(x))
exact_solution = -np.sin(x_diff)

# plot solution
plt.figure(figsize =(10,5)) #lebar 12, tinggi gambar 8
plt.plot(x_diff, foward_diff, '--',
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution,
         label = 'Exact solution/nilai turunannya')
plt.legend() # tampilkan legenda grafik
plt.show() #tampilkan grafik

# compute max error between
# numerical derivative and exat solution
max_error = max(abs(exact_solution - foward_diff))
print(max_error)
```

```python
#2. f(x) = 2 sin(3x+5)
#         = 3.2 cos(3x+5) = 6. cos (3x+5)

h=0.2

x= np.arange(0, 3*np.pi, h) #arrange = 0 - 2phi -> pi yang dipakai adalah 3.14

# compute function
y= 2*np.sin(3*x+5)

foward_diff = np.diff(y)/h

# compute corresponding grid
x_diff = x[:-1:] #

# compute exact solution (merupakan hasil turunan dari fungsi cos(x))
exact_solution = 6*np.cos(3*x_diff +5)

# plot solution
plt.figure(figsize =(10,6)) #lebar 12, tinggi gambar 8
plt.plot(x_diff, foward_diff, '--',
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution,
         label = 'Exact solution/nilai turunannya')
plt.legend() # tampilkan legenda grafik
plt.show() #tampilkan grafik

# compute max error between
# numerical derivative and exat solution
max_error = max(abs(exact_solution - foward_diff))
print(max_error)
```

```python
#3.) f(x)= 2 sin(x) + 3 cos(x)= 2 cos(x) - 3 sin(x)

h=0.2

x= np.arange(0, 3*np.pi, h) #arrange = 0 - 2phi -> pi yang dipakai adalah 3.14

# compute function
y= 2*np.sin(x) + 3*np.cos(x)

foward_diff = np.diff(y)/h

# compute corresponding grid
x_diff = x[:-1:] #

# compute exact solution (merupakan hasil turunan dari fungsi cos(x))
exact_solution = 2*np.sin(x_diff) - 3*np.sin(x_diff)

# plot solution
plt.figure(figsize =(10,6)) #lebar 12, tinggi gambar 8
plt.plot(x_diff, foward_diff, '--',
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution,
         label = 'Exact solution/nilai turunannya')
plt.legend() # tampilkan legenda grafik
plt.show() #tampilkan grafik

# compute max error between
# numerical derivative and exat solution
max_error = max(abs(exact_solution - foward_diff))
print(max_error)
```

```python
# 4.) f(x) = sin^3(2x-1) = 6 sin^2(2x-1). cos(2x-1)
# h = 0.2 dan arrange = 0 - 3pi

h=0.2

x= np.arange(0, 3*np.pi, h) #arrange = 0 - 2phi -> pi yang dipakai adalah 3.14

# compute function
y= np.sin(2*x-1)**3

foward_diff = np.diff(y)/h

# compute corresponding grid
x_diff = x[:-1:] #

# compute exact solution (merupakan hasil turunan dari fungsi cos(x))
exact_solution = 6*(np.sin(2*x_diff-1))**2 * np.cos(2*x_diff-1)

# plot solution
plt.figure(figsize =(10,6)) #lebar 12, tinggi gambar 8
plt.plot(x_diff, foward_diff, '--',
         label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution,
         label = 'Exact solution/nilai turunannya')
plt.legend() # tampilkan legenda grafik
plt.show() #tampilkan grafik

# compute max error between
# numerical derivative and exat solution
max_error = max(abs(exact_solution - foward_diff))
print(max_error)
```

COMPUTING INTEGRALS IN PYTHON

```python
# RIEMANN INTEGRAL

import numpy as np

#batas bawah = 0 = a
a=0

#batas atas = pi
b=np.pi

#n = index max
n = 11


h=(b-a) / (n-1)
x = np.linspace(a,b,n)
f = np.sin(x)

I_riemannL = h * sum(f[:n-1])
err_riemannL = 2 - I_riemannL

I_riemannR= h * sum(f[1::])
err_riemannR = 2 - I_riemannR

I_mid = h * sum(np.sin((x[:n-1] + x[1:])/2))
err_mid = 2 - I_mid

print(I_riemannL)
print(err_riemannL)

print(I_riemannR)
print(err_riemannR)

print(I_mid)
print(err_mid)
```

**TRAPEZOID RULE**


```python
# TRAPEZOID RULE

import numpy as np

#batas bawah
a = 0

#batas atas
b= np.pi

#
n = 11
h= (b-a) /(n-1)
x = np.linspace(a,b,n)
f = np.sin(x)

I_trap = (h/2) * (f[0] + 2 * sum(f[1:n-1]) + f[n-1])

err_trap = 2 - I_trap

print(I_trap)
print(err_trap)
```

```python
import numpy as np
from scipy.integrate import trapz

a=0
b= np.pi
n=11
h=(b-a) / (n-1)
x= np.linspace(a,b,n)
f = np.sin(x)

I_trapz = trapz(f,x)
I_trap = (h/2) *(f[0] + 2 * sum(f[1:n-1]) + f[n-1])

print(I_trapz)
print(I_trap)

```
