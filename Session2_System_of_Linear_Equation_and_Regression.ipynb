**step #1:**CHECK COEFFICIENT MATRIX DIAGONALLY DOMINANT OR NOT

```python
a= [[8,3,-3], [-2, -8, 5], [3,5,10]]

# sesuai baris 1, baris 2, baris 3
```

```python
# find diagnoal coefficients

# ubah matrix menjadi matriks diagonal
# a=8, b=8, c = 10

diag = np.diag(np.abs(a))
print(diag)
```

```python
# find row sum tanpa diagonal

# tiap baris dijumlahkan dan dikurangi diag
off_diag = np.sum(np.abs(a), axis=1) - diag
print(off_diag)
```

```python
if np.all(diag > off_diag):
  print('matrix is diagonally dominant')
else:
  print('NOT diagonally dominant')

  # semua var abc diag lebih besar dr off_diag,
  # tidak dominan jk : ada yg < off_diag
```

**#2** GAUSS SEIDEL METHOD

```python
x1=0
x2=0
x3=0
epsilon = 0.002
converged= False

x_old = np.array([x1,x2,x3])

x_old

# epsilon = nilai batas minimum
```

**CREATE YOUR OWN GAUSS SEIDEL SOLVER**

```python
print('Iteration results')
print(' k, x1, x2, x3 ')
for k in range(1,50):
  x1 = (14-3*x2+3*x3)/8
  x2=(5+2*x1-5*x3)/(-8)
  x3=(-8-3*x1-5*x2)/(10)
  x = np.array([x1,x2,x3])

  # awalnya substitusi x1=x2=x3=0 kedalam rumus
  # --> hasil dimasukan ke iterasi pertama

  # check if it is smaller than threshold
  dx=np.sqrt(np.dot(x-x_old, x-x_old))

  print("%d, %.4f, %.4f, %.4f"% (k, x1,x2,x3))
  if dx < epsilon:
    converged= True
    print('Converged!')
    break

  # assign the latest x value to the old value
  x_old = x;

if not converged:
  print('Not converge, increase the # of the iterations')

  # tdk converged jk nilai akhir perhitungan -> nilai x3 selisih dgn sebelumnya jauh(tidak 0,002)
```

```python
# use numpy.linalg.solve

import numpy as np

A= np.array([[4,3,-5],
  [-2,-4,5],
   [8,8,0]])
y= np.array([2,5,-3])

x=np.linalg.solve(A,y)
print(x)
```

```python
A_inv = np.linalg.inv(A)

x= np.dot(A_inv, y)
print(x)
```

**LEAST SQUARES REGRESSION IN PHYTON**




```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

# generate x and y
x=np.linspace(0,1,101)
y=1+x+x * np.random.random(len(x))

# assemble matrix A
A=np.vstack([x, np.ones(len(x))]).T

# turn y into a column vector
y=y[:, np.newaxis]

# direct least square regression
alpha=np.dot((np.dot(np.linalg.inv(np.dot(A.T, A)),A.T)), y)

print(alpha)
```

```python
# USE DIRECT INVERSE METHOD

# plot the results
plt.figure(figsize = (10,8))
plt.plot(x,y, 'b.')
plt.plot(x,alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show
```

```python
pinv= np.linalg.pinv(A)
alpha = pinv.dot(y)
print(alpha)
```

```python
alpha = np.linalg.lstsq(A, y, rcond=None)[0]
alpha
```

```python
# use optimize.curve_fit from scipy
x=np.linspace(0,1,101)
y= 1+x+x * np.random.random(len(x))

def func(x,a,b):
  y=a*x+b
  return y;

alpha = optimize.curve_fit(func, xdata=x, ydata=y)[0]
alpha
```
